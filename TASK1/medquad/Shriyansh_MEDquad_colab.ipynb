```python
import json
from transformers import pipeline, AutoTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the Medquad qna data after preprocessing from the JSON file
# Number of tokens: 5940272
with open('/content/medquad_onlyqna.json') as f:
    faq_data = json.load(f)

questions = [item['question'] for item in faq_data['data']]
contexts = [item['answer'] for item in faq_data['data']]

vectorizer = TfidfVectorizer()
context_vectors = vectorizer.fit_transform(contexts)
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)

def custom_retrieve(query, vectorizer, context_vectors, contexts, k=5):
    query_vector = vectorizer.transform([query])
    similarities = cosine_similarity(query_vector, context_vectors).flatten()
    top_indices = similarities.argsort()[-k:][::-1]
    retrieved_contexts = [contexts[idx] for idx in top_indices]
    return retrieved_contexts

def format_prompt(question, contexts):
    template = ("""system
You are a knowledgeable and helpful AI assistant capable of answering questions on a wide range of topics. Your goal is to provide accurate, relevant, and comprehensive information to users based on their queries. If you do not have enough information to fully answer a question, you should politely acknowledge the limitations in your knowledge and provide any relevant insights you can.
assistant.
If you don't know the answer to a question your answer should be I do not have the information about your question.
Question: {ques}
Answer: {ans}
""")
    context = " ".join(contexts)
    return template.format(ques=question, ans=context)
def get_response(question, contexts):
    prompt = format_prompt(question, contexts)
    sequences = text_generator(prompt, max_length=128, num_return_sequences=1)
    gen_text = sequences[0]['generated_text']
    return gen_text.split("Answer:")[-1].strip()
query = "Was ist die Behandlung für Morbus Kawasaki?" # german for What is the treatment for Kawasaki's disease

top_contexts = custom_retrieve(query, vectorizer, context_vectors, contexts) # retrival of top contexts
response = get_response(query, top_contexts)
print(response)
```

```python
#app can be used if gpu enabled, will be saved as app.py in your current dir
%%writefile app.py
pip install accelerate==0.29.3 bitsandbytes==0.43.1 transformers==4.40.0 langchain einops scipy faiss-cpu llama-index llama_hub streamlit scikit-learn numpy langdetect

import streamlit as st
import json
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import faiss
import json
import torch
from transformers import (AutoTokenizer,
                          AutoModelForCausalLM,
                          BitsAndBytesConfig,
                          TextStreamer,
                          pipeline)



HF_TOKEN ="" #insert token here
# OR:
#create a file 'environ' which has the empty (HF_TOKEN) token
#save a secret referencing HF_TOKEN'
#HF_TOKEN=os.environ.get("HF_TOKEN")

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    llm_int8_enable_fp32_cpu_offload=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)


model_name = "meta-llama/Meta-Llama-3-8B"
tokenizer= AutoTokenizer.from_pretrained(model_name,token=HF_TOKEN)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config,
    token=HF_TOKEN,

)

text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)

with open('medquad_onlyqna.json') as f:
    faq_data = json.load(f)
questions = [item['question'] for item in faq_data['data']]
contexts = [item['answer'] for item in faq_data['data']]
vectorizer = TfidfVectorizer()
context_vectors = vectorizer.fit_transform(questions)


text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)

def custom_retrieve(query, vectorizer, context_vectors, contexts, k=5):
    query_vector = vectorizer.transform([query])
    # Compute cosine similarities between the query and all context vectors
    similarities = cosine_similarity(query_vector, context_vectors).flatten()
    top_indices = similarities.argsort()[-k:][::-1]

    # Retrieve the top k most similar contexts
    retrieved_contexts = [contexts[idx] for idx in top_indices]
    return retrieved_contexts

def format_prompt(question, contexts):
    template = ("""
system
You are a knowledgeable and helpful AI assistant capable of answering questions on a wide range of topics. Your goal is to provide accurate, relevant, and comprehensive information to users based on their queries. If you do not have enough information to fully answer a question, you should politely acknowledge the limitations in your knowledge and provide any relevant insights you can.
assistant. If you don't know the answer to a question your answer should be " I dont know, can you ask me something else ?"
Question: {ques}
Answer: {ans}
""")
    context = " ".join(contexts)
    return template.format(ques=question, ans=context)

def get_response(question, contexts):
    prompt = format_prompt(question, contexts)
    # Generate text with a limited output length
    sequences = text_generator(prompt, max_length=128, num_return_sequences=1)
    gen_text = sequences[0]['generated_text']
    return gen_text.split("Answer:")[-1].strip()

# Streamlit UI
st.title("FAQ Answer Generator")

query = st.text_input("Enter your question:")

if st.button("Get Answer"):
    if query:
        # Retrieve relevant contexts
        top_contexts = custom_retrieve(query, vectorizer, context_vectors, contexts)

        # Generate a response based on the retrieved contexts
        response = get_response(query, top_contexts)

        st.write("**Answer:**")
        st.write(response)
    else:
        st.write("Please enter a question.")
```
